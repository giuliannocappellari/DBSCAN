```python
from dbscan import DBSCAN
import numpy as np
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Data/wine-clustering.csv")
scaler = StandardScaler()
X = scaler.fit_transform(df)
df["target"] = load_wine()["target"]
df
```

```python
X = df.values
X
```

```python
better = {"params":(1,2), "values":10000}
for j in range(3,20):
    for i in range(3,20):
        dbs = DBSCAN(X, i, j)
        dbs.fit()
        if abs(len(set(dbs.data_clusters.values())) - 3) * len([c for c in dbs.data_clusters.values() if c == -1]) < better["values"]:
            better["params"] = (i,j)
            better["values"] = abs(len(set(dbs.data_clusters.values())) - 3) * len([c for c in dbs.data_clusters.values() if c == -1])

better
    
```

```python
dbs = DBSCAN(X, 7, 20)
dbs.fit()
```

```python
len(set(dbs.data_clusters.values()))
```

```python
import plotly.express as px

pca = PCA()
components = pca.fit_transform(df[[col for col in df.columns if col != "target"]])
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(2),
    color=dbs.data_clusters.values()
)
fig.update_traces(diagonal_visible=False)
fig.show()
```
